````python
ts = pd.Timestamp('2020-06-06 00:00:00')
print(ts.day_name())
# 'Saturday'

# 定义一个工作小时偏移，默认是周一到周五 9-17 点，我们从 10点开始
offset = pd.offsets.BusinessHour(start='10:00')
print(offset)
# 向前偏移一个工作小时，是一个周一，跳过了周日
print(offset.rollforward(ts))
# Timestamp('2020-06-08 10:00:00')

# 向前偏移至最近的工作日，小时也会增加
print(ts + offset)
# Timestamp('2020-06-08 11:00:00')

# 向后偏移，会在周五下班前的一个小时
print(offset.rollback(ts))
# Timestamp('2020-06-05 17:00:00')

print(ts - pd.offsets.Day(1)) # 昨日
print(ts - pd.offsets.Day(2)) # 前日
print(ts - pd.offsets.Week(weekday=0) - pd.offsets.Day(14)) # 上周一
print(ts - pd.offsets.MonthEnd() - pd.offsets.MonthBegin()) # 上月一日
````
