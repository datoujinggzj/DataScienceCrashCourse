## DAY11 五大函数

```python
import pandas as pd
import numpy as np
```

```python
url = 'https://raw.githubusercontent.com/datoujinggzj/DataScienceCrashCourse/master/data/Iris.csv'
df = pd.read_csv(url)
feature_cols = df.columns[1:5]
```

### agg

```python
# 计算每个类别的四个特征的最大值。
df.groupby('Species')[feature_cols].agg(max)
```

```python
df.groupby('Species')[feature_cols].agg('max')
```

```python
df.groupby('Species')[feature_cols].agg([np.max,'min'])
```

```python
df.groupby('Species')[feature_cols].agg({
    'SepalLengthCm':min,
    'SepalWidthCm':max,
    'PetalLengthCm':np.sum
})
```

```python
df['SepalLengthCm'].agg(['max','min'])
```

```python
df[feature_cols].agg(['max','min'])
```

```python
# df.groupby('Species')[feature_cols].agg(lambda x: x.sum()-x)
# agg不可以这样用！
```

### map

```python
s = df['Species']
```

```python
s.map({
    'Iris-setosa':1
})
```

```python
s.map({
    'Iris-setosa':1,
    'Iris-versicolor':2,
    'Iris-virginica':3
})
```

```python
s.map("The species is {}".format)
```

```python
s.map(lambda x: '以a结尾' if x.endswith('a') else '不是以a结尾')
```

```python
s.map(lambda x: x[0])
```

### apply

```python
df[feature_cols].apply(np.sum)
```

```python
df[feature_cols].apply(sum,axis=1)
```

```python
def normalize(s):
    return (s-s.min())/(s.max()-s.min())
```

```python
df[feature_cols].apply(normalize)
```

```python
df.groupby('Species')[feature_cols].apply(lambda group: group.sample(5,random_state=0))
```

```python
df
```

```python
df[feature_cols].agg(lambda x: x+1)
```

```python
df[feature_cols].apply(lambda x: x+1)
```

```python
df.groupby('Species')[feature_cols].apply(lambda group: group+1)
```

```python
# df.groupby('Species')[feature_cols].agg(lambda group: group+1) 报错
```

```python
s.apply(lambda x: x[0])
```

### applymap

```python
# 每个元素转化为str
df[feature_cols].applymap(str).dtypes
```

```python
# 保留小数点后两位
df[feature_cols].applymap(lambda x:"%.2f" % x)
```

### transform

```python
df[feature_cols].transform(np.sqrt)
```

```python
df[feature_cols].transform({
    'SepalLengthCm':np.sqrt,
    'PetalWidthCm': np.exp,
})
```

```python
df.groupby('Species')[feature_cols].agg(sum)
```

```python
df.groupby('Species')[feature_cols].transform(np.mean)
```

```python
# z-score
df.groupby('Species')[feature_cols].transform(lambda x: (x-x.mean())/x.std())
```
